```python
import zipfile
import pandas as pd
```

```python
#zip to unzip file
zip_file=zipfile.ZipFile(r"C:\Users\user\orders.csv.zip")
zip_file.extractall()
zip_file.close()
```

```python
#read csv file
df=pd.read_csv (r"C:\Users\user\orders.csv.zip")#Here we have stored our dataset in variable name df
df
```

```python
#Detect missing values
df.isnull()
```

```python
#check the nan values
#we can see that there are missing values in shipmode
df.isnull().sum()
```

```python
df
```

```python
#drop nan values
df.dropna(subset=['Ship Mode'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#Renaming Columns
df.columns = df.columns.str.strip() 
df.columns = df.columns.str.lower() 
df.columns = df.columns.str.replace(' ', '_')
df.rename(columns={'Order Id':'order_id','Order Date':'order_date','Ship Mode':'ship_mode','Postal Code':'postal_code','Sub Category':'sub_category','Product Id':'product_id','cost price':'cost_price','List Price':'list_Price','Discount Percent':'discount_Percent'}, inplace=True)
```

```python
#returns top 5 rows of a dataframe
df.head()
```

```python
#returns last 5 of a dataframe
df.tail()
```

```python
#listing down the columns that are there in our dataset
print(df.columns)
```

```python
#we use describe() method to better understand the numeric data.
df.describe()
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

print(df)
```

```python
df['ship_mode'].unique()
```

```python
#CALCULATING DISCOUNTS--derive new columns discount
df['discount']=df['list_price']*df['discount_percent']/100
df
```

```python
#CALCULATING sales price--derive new columns sale_price
df['sale_price']=df['list_price']-df['discount']/100
df
```

```python
#CALCULATING profit--derive new columns profit
df['profit'] = df['list_price'] - df['cost_price']
df
```
